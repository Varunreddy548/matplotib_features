```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

categories = ['Pen','Pencil','Box']
Value1 = [3,5,6]
Value2 = [2,6,2]
Value3 = [5,3,7]

df = pd.DataFrame({
    'Pen': Value1,'Pencil':Value2,'box':Value3
},index=categories)

# Create the Stack Bar
df.plot(kind='bar',stacked=True,figsize=(8,6),color=['red','blue','green'])

plt.title('Stacked Plot Bar')
plt.xlabel('categories')
plt.ylabel('values')

plt.legend(title='Groups')

plt.show()
```

```python
# One-Hot Encoding (Module)
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# sample data
data = {
    'color' : ['red','green','blue','red','green','blue','yelloe','red','yellow']
}
df = pd.DataFrame(data)

# One - Hot Encoder using Pandas
df_encoder = pd.get_dummies(df)

df_int = df_encoder.astype(int)
print(df_int)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# sample data
data = {
    'customer_id': [1, 2, 3, 4, 5],
    'gender': ['male', 'female', 'female', 'male', 'female'],
    'region': ['north', 'south', 'east', 'west', 'north'],
    'purchase': ['yes', 'no', 'yes', 'no', 'yes']
}
df = pd.DataFrame(data)

df_encoder = pd.get_dummies(df, columns = ['gender','region','purchase'])

df_int = df_encoder.astype(int)

print(df_int)
```

```python

Sample_data = {
    'product_id': [1, 2, 3, 4, 5],
    'price': [1200, 1500, 800, 2000, 1600],
    'weight': [0.5, 1.0, 0.3, 2.0, 1.5],
    'rating': [4.5, 4.0, 5.0, 3.5, 4.2]
}
df = pd.DataFrame(Sample_data)

df_encoder = pd.get_dummies(df, columns = ['price','weight','rating'])

df_int = df_encoder.astype(int)

print(df_int)
```
