```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
# Create a line plot
plt.plot(x, y, label='sine wave')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#creating a bar chart by dictinories and key values
import matplotlib.pyplot as plt

fruits = {'Apples': 40, 'Bananas': 30, 'Kiwi': 50}
fruit_names = list(fruits.keys())  # Get the fruit names
counts = list(fruits.values())  # Get the fruit counts

# Create a bar chart
plt.bar(fruit_names, counts)
plt.xlabel('Fruits')
plt.ylabel('Count')
plt.title('Fruit Counts')
plt.show()


```
